#### Z-score for detecting outliers on int and float features

```python
zscores = {}
for col in col_int_float:
    z = np.abs(scy.stats.zscore(df[col]))
    zscores[col] = pd.Series(z)
```

```python
zscores = pd.DataFrame.from_dict(zscores)
```

```python
zscores
```

#### Quantity of outliers per column

```python
zscores_per_col = {}
for col in df[col_int_float].columns:
    #print(col, " ", zscores[zscores[col]>3].value_counts().sum())
    zscores_per_col[col] = zscores[zscores[col]>3].value_counts().sum()


zscores_per_col = pd.DataFrame.from_dict([zscores_per_col])
zscores_per_col.T
```

```python
# functions for % annots. above barplots
def sin_hue(ax, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.3f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)

# functions for % annots. in horizontal barplots
def sin_hue_h(ax, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.3f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)

def num_feat(col):
    """Simple statistics measures, Mean & Std"""
    print("Mean ", np.mean(df.iloc[:,col]))
    print("Std ", np.std(df.iloc[:,col]))
    print("Dtype ", df[df.columns[col]].dtype)

def cat_feat(col):
    """Simple statistics measures"""
    print("Frecuency ", "\n", df.iloc[:,col].value_counts()[:10])
    print("Dtype ", df[df.columns[col]].dtype)
```

```python
def plot4_num(col):
    fig = plt.figure(figsize = (12,10))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    
    # 1st
    ax = fig.add_subplot(2, 2, 1)
    sns.lineplot(x=df.iloc[:,col].index, y=df.iloc[:,col].values, ax=ax)
    plt.title(df.columns[0])
    
    #2nd
    ax = fig.add_subplot(2, 2, 2)
    pd.plotting.lag_plot(df.iloc[:,col], lag=1, ax=ax)
    plt.title(str(df.columns[col]) + " lag plot")
    
    #3rd
    ax = fig.add_subplot(2, 2, 3)
    sns.histplot(x=df.iloc[:,col], palette="pastel", ax=ax)
    
    
    plt.title(str(df.columns[col]) + " distribution")
    
    #4th
    ax = fig.add_subplot(2, 2, 4)
    scpy.stats.probplot(df.iloc[:, col], plot=ax)
    plt.title("Probability Plot of " + df.columns[col])
    
    plt.show()
```

```python
# "adr" have 0s value, so we cant do BoxCox transformation
#   then we have boxcox1p, but we need to calculate lambda parameter
#   for lambda, we can use PowerTransformer to find out the lambda value
#   and then use that value with boxcox1p for the transformation i need.

from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')
```

```python
pt.fit(df[["adr"]])
```

```python
# the optimal lambda value

print(pt.lambdas_)
```

```python
# boxcox1p transformation using the optimal lambda value, and the skew of that transformation

scpy.special.boxcox1p(df["adr"], pt.lambdas_).skew()
```

From a 0.67 to 0.048, the skew have nearly delete from the feature.

```python
# here a look on the distribution of the data after boxcox1p transformation

plt.figure(figsize=(15,8))
plt.title(df.columns[27])
sns.histplot(np.sqrt(scpy.special.boxcox1p(df["adr"], pt.lambdas_)))
```

```python
df["adr"] = scpy.special.boxcox1p(df["adr"], pt.lambdas_)
```

#### Chi-Square test

```python
crosstab1 # we need to make a crosstab or "contingency tab" for chi-square test using scipy chi2_contingency
```

```python
chivalue, pvalue, dof, expected = scpy.stats.chi2_contingency(crosstab1)
```

```python
print('Chi-Square value= %.3f, P-value= %.3f, Degree of fredoom= %.f' % (chivalue, pvalue, dof))
```

```python
significance = 0.95
critical = scpy.stats.chi2.ppf(significance, dof)
print('Significance= %.3f, Critical(Chi-Sq prob)= %.3f, Chi-Square Value= %.3f' % (significance, critical, chivalue))
```

```python
if abs(chivalue) >= critical:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```

```python
alpha = 1.0 - significance
print('Alpha= %.2f, P-value= %.2f' % (alpha, pvalue))
if pvalue <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```

Chi-square test shows that there is statistical dependence between both variables.

```python
def chi2_test(crosstab, significance):
    """
    Function to make chi2 contingency test and hypotesis test with chi2
    Parameters:
        crosstab = Contingency table, using pandas pd.crosstab
        significance = significance level for the hipotesis test, 0.00 format.
    """
    chivalue, pvalue, dof, expected = scpy.stats.chi2_contingency(crosstab)
    print('Chi-Square value= %.3f, P-value= %.3f, Degree of fredoom= %.f' % (chivalue, pvalue, dof), '\n')
    critical = scpy.stats.chi2.ppf(significance, dof)
    print('Significance= %.3f, Critical(Chi-Sq prob)= %.3f, Stat-Chi-Square Value= %.3f' % (significance, critical, chivalue), '\n')
    if abs(chivalue) >= critical:
        print(' ** According to the comparison between Chi-value and Critical-value =>', 'Dependent (reject H0)', '\n')
    else:
        print(' ** According to the comparison between Chi-value and Critical-value =>', 'Independent (fail to reject H0)', '\n')
    alpha = 1.0 - significance
    print('Alpha= %.2f, P-value= %.2f' % (alpha, pvalue), '\n')
    if pvalue <= alpha:
        print(' ** According to the comparison between P-value and Alpha =>', 'Dependent (reject H0)', '\n')
    else:
        print(' ** According to the comparison between P-value and Alpha =>', 'Independent (fail to reject H0)', '\n')
```

```python
chi2_test(crosstab1, 0.95)
```

Chi-Square test indicates that the features are dependent on each other.
